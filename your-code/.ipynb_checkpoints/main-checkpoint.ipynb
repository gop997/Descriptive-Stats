## Challenge 1
#### 1.- Define a function that simulates rolling a dice 10 times. Save the information in a dataframe.
**Hint**: you can use the *choices* function from module *random* to help you with the simulation.

```python
# Function to simulate rolling a dice 10 times
def roll_dice(n=10):
    return pd.DataFrame({"Rolls": random.choices(range(1, 7), k=n)})

# Simulate rolling the dice
dice_df = roll_dice()
dice_df
```

#### 2.- Plot the results sorted by value.

```python
import matplotlib.pyplot as plt

# Sort the dataframe and plot
dice_df_sorted = dice_df.sort_values(by="Rolls").reset_index(drop=True)
plt.plot(dice_df_sorted.index, dice_df_sorted["Rolls"], marker='o')
plt.title("Dice Rolls Sorted by Value")
plt.xlabel("Roll Number")
plt.ylabel("Dice Value")
plt.show()
```

#### 3.- Calculate the frequency distribution and plot it. What is the relation between this plot and the plot above? Describe it with words.

```python
# Calculate frequency distribution
freq_dist = dice_df["Rolls"].value_counts().sort_index()

# Plot frequency distribution
freq_dist.plot(kind="bar")
plt.title("Frequency Distribution of Dice Rolls")
plt.xlabel("Dice Value")
plt.ylabel("Frequency")
plt.show()
```

```python
"""
The frequency distribution plot shows how often each dice value appears in the 10 rolls. In comparison, the sorted plot in step 2 displays the actual dice values in ascending order, providing a visual representation of the distribution.
"""
```

## Challenge 2
Now, using the dice results obtained in *challenge 1*, your are going to define some functions that will help you calculate the mean of your data in two different ways, the median and the four quartiles. 

#### 1.- Define a function that computes the mean by summing all the observations and dividing by the total number of observations. You are not allowed to use any methods or functions that directly calculate the mean value. 

```python
def calculate_mean(data):
    total_sum = sum(data)
    count = len(data)
    return total_sum / count

# Calculate mean of dice rolls
mean_value = calculate_mean(dice_df["Rolls"])
mean_value
```

#### 2.- First, calculate the frequency distribution. Then, calculate the mean using the values of the frequency distribution you've just computed. You are not allowed to use any methods or functions that directly calculate the mean value. 

```python
# Calculate mean using frequency distribution
total_sum_freq = sum(value * count for value, count in freq_dist.items())
count_freq = sum(freq_dist)
mean_value_freq = total_sum_freq / count_freq
mean_value_freq
```

#### 3.- Define a function to calculate the median. You are not allowed to use any methods or functions that directly calculate the median value. 
**Hint**: you might need to define two computation cases depending on the number of observations used to calculate the median.

```python
def calculate_median(data):
    sorted_data = sorted(data)
    n = len(sorted_data)
    mid = n // 2
    if n % 2 == 0:
        median = (sorted_data[mid - 1] + sorted_data[mid]) / 2
    else:
        median = sorted_data[mid]
    return median

# Calculate median of dice rolls
median_value = calculate_median(dice_df["Rolls"])
median_value
```

#### 4.- Define a function to calculate the four quartiles. You can use the function you defined above to compute the median but you are not allowed to use any methods or functions that directly calculate the quartiles. 

```python
def calculate_quartiles(data):
    sorted_data = sorted(data)
    n = len(sorted_data)
    q2 = calculate_median(sorted_data)
    if n % 2 == 0:
        lower_half = sorted_data[:n//2]
        upper_half = sorted_data[n//2:]
    else:
        lower_half = sorted_data[:n//2]
        upper_half = sorted_data[n//2+1:]
    q1 = calculate_median(lower_half)
    q3 = calculate_median(upper_half)
    return q1, q2, q3

# Calculate quartiles of dice rolls
quartiles = calculate_quartiles(dice_df["Rolls"])
quartiles
```

## Challenge 4
In the `data` folder of this repository you will find three different files with the prefix `ages_population`. These files contain information about a poll answered by a thousand people regarding their age. Each file corresponds to the poll answers in different neighbourhoods of Barcelona.

#### 1.- Read the file `ages_population.csv`. Calculate the frequency distribution and plot it as we did during the lesson. Try to guess the range in which the mean and the standard deviation will be by looking at the plot. 

```python
# Read the file ages_population.csv
df_ages1 = pd.read_csv('/Users/gopikagowthaman/Desktop/IRONHACK/week03/Descriptive-Stats/data/ages_population.csv')

# Calculate the frequency distribution
age_counts = df_ages1['observation'].value_counts().sort_index()

# Plot the frequency distribution
plt.figure(figsize=(10, 6))
age_counts.plot(kind='bar')
plt.title('Frequency Distribution of Ages in the First Neighbourhood')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Provide a guess for the range of mean and standard deviation by looking at the plot
mean_guess = (30, 40)
std_dev_guess = (10, 15)

print("Guessed range for Mean:", mean_guess)
print("Guessed range for Standard Deviation:", std_dev_guess)
```

#### 2.- Calculate the exact mean and standard deviation and compare them with your guesses. Do they fall inside the ranges you guessed?

```python
# Calculate the frequency distribution
age_counts = df_ages1['observation'].value_counts().sort_index()

# Plot the frequency distribution
plt.figure(figsize=(10, 6))
age_counts.plot(kind='bar')
plt.title('Frequency Distribution of Ages in the First Neighbourhood')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Provide a guess for the range of mean and standard deviation by looking at the plot
mean_guess = (30, 40)
std_dev_guess = (10, 15)

print("Guessed range for Mean:", mean_guess)
print("Guessed range for Standard Deviation:", std_dev_guess)

# Calculate the exact mean and standard deviation
exact_mean = df_ages1['observation'].mean()
exact_std_dev = df_ages1['observation'].std()

print("Exact Mean:", exact_mean)
print("Exact Standard Deviation:", exact_std_dev)

# Compare guesses with actual values
mean_in_range = mean_guess[0] <= exact_mean <= mean_guess[1]
std_dev_in_range = std_dev_guess[0] <= exact_std_dev <= std_dev_guess[1]

print("Is the exact mean within the guessed range?", mean_in_range)
print("Is the exact standard deviation within the guessed range?", std_dev_in_range)
```

```python
"""
Guessed range for Mean: (30, 40)
Guessed range for Standard Deviation: (10, 15)
Exact Mean: 36.56
Exact Standard Deviation: 12.81649962597677
Is the exact mean within the guessed range? True
Is the exact standard deviation within the guessed range? True
"""
```

#### 5.- Calculate the mean and standard deviation. Compare the results with the mean and standard deviation in step 2. What do you think?

```python
# Calculate the exact mean and standard deviation for the first dataset
exact_mean_1 = df_ages1['observation'].mean()
exact_std_dev_1 = df_ages1['observation'].std()

print("Exact Mean for First Neighbourhood:", exact_mean_1)
print("Exact Standard Deviation for First Neighbourhood:", exact_std_dev_1)

# Calculate the exact mean and standard deviation for the second dataset
exact_mean_2 = df_ages2['observation'].mean()
exact_std_dev_2 = df_ages2['observation'].std()

print("Exact Mean for Second Neighbourhood:", exact_mean_2)
print("Exact Standard Deviation for Second Neighbourhood:", exact_std_dev_2)

# Plot histograms for both datasets
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(df_ages1['observation'], bins=range(0, 85, 5), edgecolor='black')
plt.title('Histogram of Ages in the First Neighbourhood')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(df_ages2['observation'], bins=range(0, 85, 5), edgecolor='black')
plt.title('Histogram of Ages in the Second Neighbourhood')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Compare the results
print("Comparison of Means:")
print("First Neighbourhood Mean:", exact_mean_1)
print("Second Neighbourhood Mean:", exact_mean_2)

print("\nComparison of Standard Deviations:")
print("First Neighbourhood Standard Deviation:", exact_std_dev_1)
print("Second Neighbourhood Standard Deviation:", exact_std_dev_2)
```

#### 2.- Calculate the mean and standard deviation. Compare the results with the plot in step 1. What is happening?

```python
# Calculate the exact mean and standard deviation for the third dataset
exact_mean_3 = df_ages3['observation'].mean()
exact_std_dev_3 = df_ages3['observation'].std()

print("Exact Mean for Third Neighbourhood:", exact_mean_3)
print("Exact Standard Deviation for Third Neighbourhood:", exact_std_dev_3)

# Compare the results with the plot in step 1
print("Comparison of Means:")
print("First Neighbourhood Mean:", exact_mean_1)
print("Second Neighbourhood Mean:", exact_mean_2)
print("Third Neighbourhood Mean:", exact_mean_3)

print("\nComparison of Standard Deviations:")
print("First Neighbourhood Standard Deviation:", exact_std_dev_1)
print("Second Neighbourhood Standard Deviation:", exact_std_dev_2)
print("Third Neighbourhood Standard Deviation:", exact_std_dev_3)
```

#### 3.- Calculate the four quartiles. Use the results to explain your reasoning for question in step 2. How much of a difference is there between the median and the mean?

```python
# Calculate the four quartiles
quartiles = df_ages3['observation'].quantile([0.25, 0.5, 0.75, 1.0])
median = df_ages3['observation'].median()

print("Quartiles for Third Neighbourhood:")
print(quartiles)
print("Median for Third Neighbourhood:", median)
```

```python
# Compare the mean and the median
difference_mean_median = exact_mean_3 - median
print("Difference between Mean and Median:", difference_mean_median)

# Summary of results and interpretation
print("\nSummary:")
print(f"Mean: {exact_mean_3}")
print(f"Median: {median}")
print(f"Standard Deviation: {exact_std_dev_3}")
print(f"Quartiles: \n{quartiles}")
print(f"Difference between Mean and Median: {difference_mean_median}")
```

```python
"""The mean is higher than the median, which indicates that the distribution is slightly skewed to the right. This means there are some higher ages that are pulling the mean up. The standard deviation of 16.14 reflects a wide spread of ages, showing significant variability.
In summary, the third neighbourhood has a relatively older population with a wide age range, and the distribution is slightly right-skewed, meaning a few higher age values are increasing the mean more than the median.
"""
```
